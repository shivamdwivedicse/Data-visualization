```python
import matplotlib.pyplot as plt

data=["system","bhaukal", "aura", "bahubali"]
numbers=[12,20,33,40]
plt.title("UP WALE")
plt.xlabel("data")
plt.ylabel("numbers")
plt.bar(data,numbers)
plt.show()
```

```python
#simple line graph
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y)
plt.title("Simple line graph using matplotlib")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ["Math", "DS", "OS", "Python"]
marks = [80, 70, 85, 90]

plt.bar(subjects, marks)
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# Horizontal Bar Graph
import matplotlib.pyplot as plt
languages =["c","c++","Python","java"]
popularity =[70,65,90,75]
colors=["#264653", "#2A9D8F", "#E9C46A", "#F4A261"]
plt.barh(languages,popularity)
plt.title("Programming language Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Languages")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
shivam = ["Python", "Java", "C++", "C"]

plt.pie(sizes, labels=shivam, autopct="%1.1f%%")
plt.title("Language Usage")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.scatter(x, y)
plt.title("scatter Plot example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# histogram
import matplotlib.pyplot as plt

marks = [45, 55, 60, 65, 70, 75, 80, 85, 90]

plt.hist(marks)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 15, 25, 30]

plt.plot(x, y, marker='*')
plt.title("Line Graph with Markers")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.title("Multiple Lines")
plt.show()
```
