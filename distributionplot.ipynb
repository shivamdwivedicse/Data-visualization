```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = sns.load_dataset("tips")
df
```

```python
df["size"].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins =20 , kde = True) #histogram is used to check distribution
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins =20 , kde = True)
```

**JOINTPLOT**

```python
#used for comparisons
sns.jointplot(x="total_bill" , y="tip" , data=df , kind ="scatter")
```

```python
sns.jointplot(x="total_bill" , y="tip" , data=df ,kind="kde")
```

```python
sns.jointplot(x="total_bill" , y="tip" , data=df , kind ="hex")
```

```python
sns.jointplot(x="total_bill" , y="tip" , data=df , kind ="reg")
```

**PAIRPLOT**

```python
sns.pairplot(df,hue="sex",palette="rainbow") 
```

```python
sns.pairplot(df,hue="size",palette="rainbow")
```
