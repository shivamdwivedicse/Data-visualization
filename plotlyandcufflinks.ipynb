```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import cufflinks as cf 
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
px.line(tips['total_bill'] , color_discrete_sequence=['orange'])
```

```python
tip1=tips.groupby('day')['tip'].mean()
tip1
```

```python
px.bar(tip1)
```

```python
px.scatter(tips,x="total_bill" , y= 'tip')
```

```python
px.box(tips[['total_bill','day']],x ='total_bill',y='day')
```
