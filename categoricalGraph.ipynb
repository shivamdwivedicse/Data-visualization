```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
df = sns.load_dataset("tips")
df
```

**COUNT PLOT**

```python
 sns.countplot(x= df["sex"],hue = df['smoker'])
```

**BARPLOT**


```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

```python
sns.barplot(x=df['sex'], y=df['total_bill'],estimator=np.sum)
```

```python
sns.barplot(x=df['sex'], y=df['tip'],estimator=np.sum)
```

**BOX PLOT**

```python
#becomes by using 5 number summary
# sns.boxplot(x='tip', y='day',data = df )
sns.boxplot(x = df['tip'] , y=df['day'], palette="rainbow")
```

**VIOLINPLOT**

```python
sns.violinplot(x= 'day' , y='total_bill', data =df, palette="rainbow" )
```

**STRIP PLOT AND SWARM PLOT**

```python
sns.stripplot(x='day',y='total_bill',data =df, palette='rainbow')
```

```python
sns.swarmplot(x='day',y='total_bill',data = df , palette="rainbow")
```

```python
#combing swarmplot and violin plot:-
sns.violinplot(x= 'tip' , y='day' , data = df )
sns.swarmplot(x= 'tip' , y='day' , data = df )
```
